```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import seaborn as sns
# from seabron import load_dataset
#titanic dataset
df= pd.read_csv(r"C:\Users\sai\Downloads\Titanic (1).csv")
```

```python
# 1.CountPlot
sns.countplot(df['Survived'])
plt.show()
```

```python
# 2. Pie Chart
df['Sex'].value_counts().plot(kind="pie", autopct="%.2f")
plt.show()
```

```python
# 3. Histogram

plt.hist(df['Age'], bins=4)
plt.show()
```

```python
# Distplot 

sns.distplot(df['Survived'])
plt.show()
```

```python
# BoxPlot

sns.boxplot(df['Sex'], df['Age'])
```

```python
# Bar Plot

sns.barplot(df['Survived'], df['PClass'])
plt.show()
```

```python
# Scatter Plot

sns.scatterplot(df['Age'], df['Sex'])
```

```python
# Multivariate analysis with scatter plot

sns.scatterplot(df['PClass'], df['Age'], df['Sex'])
plt.show()
                                                       
```

```python
# Multivariate analysis with boxplot

sns.boxplot(df['Age'], df['PClass'], df['Survived'])
plt.show()
```

```python
# Heatmap

sns.heatmap(pd.crosstab(df['Sex'], df['Survived']))
```

```python
# Cluster map

sns.clustermap(pd.crosstab(df['PClass'], df['Sex']))
plt.show()
```
